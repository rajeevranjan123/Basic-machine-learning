# 4. Train Data

```python
X =  titanic_data.drop("Survived", axis=1)
Y = titanic_data['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

#  What is logistic regression
Logistic regression is a supervised learning technique in which we are classifing the binary data. 
logistic regression is used to describe the data and to explain the relationship between independent variable to dependent(binary) variable.
<br>
A logistic growth curve is an S-shaped (sigmoidal) curve that can be used to model functions that increase gradually at first, more rapidly in the middle growth period, and slowly at the end, leveling off at a maximum value after some period of time.

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train, y_train)
```

```python
pred = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test, pred)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, pred)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, pred)
```
